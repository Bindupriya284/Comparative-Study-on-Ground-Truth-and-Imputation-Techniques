```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('ETL_DATA_new.csv')
df
```

```python
# Set the index to 'TimeStamp'
df.set_index('TimeStamp', inplace=True)
df
```

```python
# Remove columns with more than 100% NaN values
threshold = 1.00
null_percentages = (df.isnull().sum() / len(df)) * 100
columns_to_remove = null_percentages[null_percentages > threshold].index
df1 = df.drop(columns=columns_to_remove)
df1
```

```python
# Remove negative values
df1[df1 < 0] = abs(df1[df1 < 0])
df1
```

```python
# Create a mask for introducing additional NaNs
nan_mask = np.random.rand(*df1.shape) < 0.1
df_with_nans = df1.mask(nan_mask)
df_with_nans
```

```python
# Save the original values at positions where NaNs were introduced
original_values = df1[nan_mask]
original_values
```

```python
# Forward fill
df_ff = df_with_nans.ffill()
df_ff
```

```python
# Backward fill
df_bf = df_with_nans.bfill()
df_bf
```

```python
# Mean fill
df_mean = df_with_nans.fillna(df_with_nans.mean())
df_mean
```

```python
# Mode fill
df_mode = df_with_nans.fillna(df_with_nans.mode().iloc[0])
df_mode
```

```python
# Median fill
df_median = df_with_nans.fillna(df_with_nans.median())
df_median
```

```python
# Combine all imputed dataframes
combined_imputed = (df_ff + df_bf + df_mean + df_mode + df_median) / 5
combined_imputed
```

```python
# Calculate accuracy
def calculate_overall_accuracy(original, imputed):
    comparison = original == imputed
    accuracy = comparison.sum().sum() / original.size
    return accuracy

# Calculate overall accuracy
overall_accuracy = calculate_overall_accuracy(df1, combined_imputed)

# Print overall accuracy
print(f'Overall Accuracy: {overall_accuracy}')
```
